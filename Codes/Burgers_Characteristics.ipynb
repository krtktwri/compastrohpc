```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x_ini = np.linspace(0.01, 1-0.01, 48)
u_ini = np.sin(2*np.pi*x_ini) #u(x, 0)

x = np.linspace(0, 1, 40)
t = ( np.outer(x,np.ones(48)) - np.outer(np.ones(40),x_ini) )/np.outer(np.ones(40),u_ini)
```

```python
plt.subplot(211)
plt.title('Burgers equation characteristics')
plt.plot(x_ini, u_ini)
plt.ylabel('u(x,0)')
plt.grid()

plt.subplot(212)
plt.plot(x, t)
plt.grid()
plt.ylim([0,0.28])
plt.xlabel('x')
plt.ylabel('t')
```
